```python
import config

# Assuming df is your DataFrame loaded with data
# Google Maps API key
Google_Maps_API_KEY = config.Google_Maps_API_KEY  # to remove/replace this later on!! 

import requests

def validate_address(address, Google_Maps_API_KEY):
    # Define the endpoint URL
    url = "https://maps.googleapis.com/maps/api/place/findplacefromtext/json"

    # Define the parameters for the API request
    params = {
        'input': address,
        'inputtype': 'textquery',
        'fields': 'formatted_address,geometry',
        'key': Google_Maps_API_KEY
    }

    # Send the GET request to the API
    response = requests.get(url, params=params)

    # If the request was successful, return the data in JSON format
    if response.status_code == 200:
        return response.json()
    else:
        return None
```

```python
# Define a function to parse JSON and extract formatted_address
def extract_formatted_address(json_str):
    data = json.loads(json_str)
    if 'candidates' in data and data['candidates']:
        return data['candidates'][0]['formatted_address']
    return None

# Define a function to parse JSON and extract latitude
def extract_latitude(json_str):
    data = json.loads(json_str)
    if 'candidates' in data and data['candidates']:
        return data['candidates'][0]['geometry']['location']['lat']
    return None

# Define a function to parse JSON and extract longitude
def extract_longitude(json_str):
    data = json.loads(json_str)
    if 'candidates' in data and data['candidates']:
        return data['candidates'][0]['geometry']['location']['lng']
    return None

# Apply the functions to the 'validatedResult' column to create new columns
filtered_df['formatted_address'] = filtered_df['validatedResult'].apply(extract_formatted_address)
filtered_df['latitude'] = filtered_df['validatedResult'].apply(extract_latitude)
filtered_df['longitude'] = filtered_df['validatedResult'].apply(extract_longitude)
```
